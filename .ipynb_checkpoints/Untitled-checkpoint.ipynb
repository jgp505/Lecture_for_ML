```python
import os
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
from sklearn.model_selection import KFold, cross_val_score
import joblib

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
df = pd.read_csv("data.csv",index_col=0)
df.dropna()

index = df.index.tolist()
np.random.shuffle(index)

ratio = int(len(index)*0.7)
train_index = index[:ratio]
test_index = index[ratio:]

train = df.loc[train_index]
test = df.loc[test_index]
```

```python
train_y = train[['log_KVRH']]
test_y = test[['log_KVRH']]

train_x = train.drop(['K_VRH','log_KVRH','pretty_formula','icsd_ids','composition'],axis=1)
test_x = test.drop(['K_VRH','log_KVRH','pretty_formula','icsd_ids','composition'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize=False)
lr.fit(train_x, train_y)
pred_y_lr = lr.predict(test_x)

mae = mean_absolute_error(test_y,pred_y_lr)
mse = mean_squared_error(test_y,pred_y_lr)
r2 = r2_score(test_y,pred_y_lr)

print("Model Validation")
print("MAE : %.3f\tMSE : %.3f\tR2 : %.3f\n\n"%(mae,mse,r2))


crossvalidation = KFold(n_splits=10, shuffle=True, random_state=1)
r2_scores = cross_val_score(lr, test_x, test_y, scoring='r2', cv=crossvalidation, n_jobs=-1)

for e,i in enumerate(r2_scores) :
    if e == 0 :
        serial = '1st'
    elif e == 1 :
        serial = '2nd' 
    elif e == 2 :
        serial = '3rd'
    else :
        serial = "%i th"%(e+1)
    print("%s R2 score : %.3f" %(serial, i))
    
    
plt.scatter(test_y,pred_y_lr,edgecolor='k')
plt.plot([0,3],[0,3],'k-',lw=.5)
plt.xlim(0,3);plt.ylim(0,3)
plt.xlabel("DFT K_VRH")
plt.ylabel("Predicted K_VRH")
plt.title("Linear Regression")
```

```python
from sklearn.ensemble import RandomForestRegressor


rf = RandomForestRegressor()
rf.fit(train_x, train_y)

pred_y_rf =rf.predict(test_x)

mae = mean_absolute_error(test_y,pred_y_rf)
mse = mean_squared_error(test_y,pred_y_rf)
r2 = r2_score(test_y,pred_y_rf)

print("Model Validation")
print("MAE : %.3f\tMSE : %.3f\tR2 : %.3f\n\n"%(mae,mse,r2))


crossvalidation = KFold(n_splits=10, shuffle=True, random_state=1)
r2_scores = cross_val_score(rf, test_x, test_y, scoring='r2', cv=crossvalidation, n_jobs=-1)

for e,i in enumerate(r2_scores) :
    if e == 0 :
        serial = '1st'
    elif e == 1 :
        serial = '2nd' 
    elif e == 2 :
        serial = '3rd'
    else :
        serial = "%i th"%(e+1)
    print("%s R2 score : %.3f" %(serial, i))
    
    
plt.scatter(test_y,pred_y_lr,edgecolor='k')
plt.plot([0,3],[0,3],'k-',lw=.5)
plt.xlim(0,3);plt.ylim(0,3)
plt.xlabel("DFT K_VRH")
plt.ylabel("Predicted K_VRH")
plt.title("Random Forest Regression")
```

```python
plt.figure(figsize=(8,14))
coef = rf.feature_importances_
coef_ser = pd.Series(rf.feature_importances_, index = train_x.columns)
imp_coef = coef_ser.sort_values()
imp_coef_002 = imp_coef[imp_coef>0.01]*100
plt.rcParams['font.size'] = 14
imp_coef_002.plot(kind = "barh",color='#6B4058')
plt.title("Feature importance using Random Forest Model", fontsize=28)
plt.xlabel("\nFeature importance (%)", fontsize=21)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(n_estimators=100,max_depth=4)
gbr.fit(train_x, train_y)

pred_y_gbr =gbr.predict(test_x)

mae = mean_absolute_error(test_y,pred_y_gbr)
mse = mean_squared_error(test_y,pred_y_gbr)
r2 = r2_score(test_y,pred_y_gbr)

print("Model Validation")
print("MAE : %.3f\tMSE : %.3f\tR2 : %.3f\n\n"%(mae,mse,r2))


crossvalidation = KFold(n_splits=10, shuffle=True, random_state=1)
r2_scores = cross_val_score(gbr, test_x, test_y, scoring='r2', cv=crossvalidation, n_jobs=-1)

for e,i in enumerate(r2_scores) :
    if e == 0 :
        serial = '1st'
    elif e == 1 :
        serial = '2nd' 
    elif e == 2 :
        serial = '3rd'
    else :
        serial = "%i th"%(e+1)
    print("%s R2 score : %.3f" %(serial, i))
    
    
plt.scatter(test_y,pred_y_lr,edgecolor='k')
plt.plot([0,3],[0,3],'k-',lw=.5)
plt.xlim(0,3);plt.ylim(0,3)
plt.xlabel("DFT K_VRH")
plt.ylabel("Predicted K_VRH")
plt.title("Gradient Bossting Regression")
```
