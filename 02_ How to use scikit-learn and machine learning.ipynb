# 선형회귀 (Linear Regression)

통계학에서 선형 회귀(線型回歸, 영어: linear regression)는 종속 변수 y와 한 개 이상의 독립 변수 (또는 설명 변수) X와의 선형 상관 관계를 모델링하는 회귀분석 기법이다. 한 개의 설명 변수에 기반한 경우에는 단순 선형 회귀(simple linear regression), 둘 이상의 설명 변수에 기반한 경우에는 다중 선형 회귀라고 한다.

In statistics, linear regression is a linear approach for modelling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables). The case of one explanatory variable is called simple linear regression; for more than one, the process is called multiple linear regression. This term is distinct from multivariate linear regression, where multiple correlated dependent variables are predicted, rather than a single scalar variable.

```python
from sklearn.linear_model import LinearRegression

array = data[['RM','Price']].values
lr = LinearRegression()
lr.fit(X=array[:,0].reshape(-1,1),y=array[:,1])
prediction = lr.predict(array[:,0].reshape(-1,1))


print("a value : ",lr.intercept_)
print("b value : ", lr.coef_[0])

print("y = %.3fx - %.3f"%(lr.coef_,lr.intercept_))

plt.scatter(data['RM'],data['Price'],color='g')
plt.plot(data['RM'],prediction,'k-')

plt.xlabel("RM")
plt.ylabel("Price")
```

```python
from sklearn.linear_model import LinearRegression

array = data[['LSTAT','Price']].values
lr = LinearRegression()
lr.fit(X=array[:,0].reshape(-1,1),y=array[:,1])
prediction = lr.predict(array[:,0].reshape(-1,1))


print("a value : ",lr.intercept_)
print("b value : ", lr.coef_[0])

print("y = %.3fx - %.3f"%(lr.coef_,lr.intercept_))

plt.scatter(data['LSTAT'],data['Price'],color='g')
plt.plot(data['LSTAT'],prediction,'k-')

plt.xlabel("LSTAT")
plt.ylabel("Price")
```

# 랜덤 포레스트 (RandomForest)

기계 학습에서의 랜덤 포레스트(영어: random forest)는 분류, 회귀 분석 등에 사용되는 앙상블 학습 방법의 일종으로, 훈련 과정에서 구성한 다수의 결정 트리로부터 부류(분류) 또는 평균 예측치(회귀 분석)를 출력함으로써 동작한다.

```python
from sklearn.ensemble import RandomForestRegressor
X = data[feature] ; Y = data['Price']

rf = RandomForestRegressor()
rf.fit(X=X,y=Y)
prediction = rf.predict(X)

fig, axes = plt.subplots(1,2)
axes[0].scatter(data['Price'],prediction,c='b')

axes[0].set_xlabel("Real Price")
axes[0].set_ylabel("Predicted Price")
axes[0].plot([0,60],[0,60],'k-')
axes[0].set_xlim(5,52) ; axes[0].set_ylim(5,52)

coef = rf.feature_importances_
coef_ser = pd.Series(rf.feature_importances_, index = feature)
imp_coef = coef_ser.sort_values()
imp_coef_002 = imp_coef[imp_coef>0.01]*100
imp_coef_002.plot(kind = "barh",color='#6B4058')
axes[1].set_title("Feature importance using \nRandom Forest Model", fontsize=14)
axes[1].set_xlabel("Feature importance (%)", fontsize=14)
fig.tight_layout()

plt.show()
```

# Gradient boosting 

Gradient boosting is a machine learning technique used in regression and classification tasks, among others. It gives a prediction model in the form of an ensemble of weak prediction models, which are typically decision trees.[1][2] When a decision tree is the weak learner, the resulting algorithm is called gradient-boosted trees; it usually outperforms random forest.[1][2][3] A gradient-boosted trees model is built in a stage-wise fashion as in other boosting methods, but it generalizes the other methods by allowing optimization of an arbitrary differentiable loss function.

```python
from sklearn.ensemble import GradientBoostingRegressor

X = data[feature] ; Y = data['Price']

gbr = GradientBoostingRegressor(n_estimators=100,max_depth=4)
gbr.fit(X=X,y=Y)
prediction = gbr.predict(X)

fig, axes = plt.subplots(1,2)
axes[0].scatter(data['Price'],prediction,c='r')

axes[0].set_xlabel("Real Price")
axes[0].set_ylabel("Predicted Price")
axes[0].plot([0,60],[0,60],'k-')
axes[0].set_xlim(5,52) ; axes[0].set_ylim(5,52)

coef = gbr.feature_importances_
coef_ser = pd.Series(gbr.feature_importances_, index = feature)
imp_coef = coef_ser.sort_values()
imp_coef_002 = imp_coef[imp_coef>0.01]*100
imp_coef_002.plot(kind = "barh",color='#6B4058')
axes[1].set_title("Feature importance using \nGradient Boosting Model", fontsize=14)
axes[1].set_xlabel("Feature importance (%)", fontsize=14)
fig.tight_layout()

plt.show()
```
